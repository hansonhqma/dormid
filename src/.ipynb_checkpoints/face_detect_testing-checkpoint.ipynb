```python
import numpy as np
import cv2 as cv
import cvlib as cb
import os
import matplotlib.pyplot as plt
import deploy
```

```python
def facewithpadding(pic, padding):
    h = pic.shape[0]
    w = pic.shape[1]
    f, _ = cb.detect_face(pic)
    if len(f) < 1:
        print("no face found")
        plt.imshow(pic)
        return pic, False
    elif len(f) > 1: # mulitple faces found, skip
        return pic, False
    f = f[0]
    pic = pic[max(0, f[1]-padding):min(h, f[3]+padding), max(0, f[0]-padding):min(w, f[2]+padding)]
    pic = cv.resize(pic, (100,125))
    return cv.cvtColor(pic, cv.COLOR_BGR2GRAY), True
```

```python
pic = cv.imread('pic2.JPG')
faces, c = cb.detect_face(pic)
faces = faces[0]
padding = 10
me = cv.cvtColor(pic[faces[1]-padding:faces[3]+padding, faces[0]-padding:faces[2]+padding], cv.COLOR_BGR2GRAY)
me = cv.resize(me, (100,125))
me = me.astype('float32')/255
plt.imshow(me)
```

```python
out, value, pair = deploy.prediction(me)
value
```
